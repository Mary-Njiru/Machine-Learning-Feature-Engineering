```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
vehicle_collisions_df=pd.read_csv('/home/oem/machine_learning_jupyter/motor_vehicle_collisions_crashes.csv')
```

```python
vehicle_collisions_df.head(1000)
```

```python
vehicle_collisions_df.rename(columns={
    'CRASH DATE': 'crash_date',
    'CRASH TIME': 'crash_time',
    'BOROUGH': 'new_york_borough',
    'ZIP CODE': 'zip_code',
    'LATITUDE': 'latitude',
    'LONGITUDE': 'longitude',
    'LOCATION': 'location',
    'ON STREET NAME': 'on_street_name',
    'CROSS STREET NAME': 'cross_street_name',
    'OFF STREET NAME': 'off_street_name',
    'NUMBER OF PERSONS INJURED': 'number_of_persons_injured',
    'NUMBER OF PERSONS KILLED': 'number_of_persons_killed',  
    'NUMBER OF PEDESTRIANS INJURED': 'number_of_pedestrians_injured',
    'NUMBER OF PEDESTRIANS KILLED': 'number_of_pedestrians_killed',
    'NUMBER OF CYCLIST INJURED': 'number_of_cyclists_injured',
    'NUMBER OF CYCLIST KILLED': 'number_of_cyclists_killed',
    'NUMBER OF MOTORIST INJURED': 'number_of_motorist_injured',
    'NUMBER OF MOTORIST KILLED': 'number_of_motorist_killed',
    'CONTRIBUTING FACTOR VEHICLE 1': 'contributing_factor_vehicle_1',
    'CONTRIBUTING FACTOR VEHICLE 2': 'contributing_factor_vehicle_2',
    'CONTRIBUTING FACTOR VEHICLE 3': 'contributing_factor_vehicle_3',
    'CONTRIBUTING FACTOR VEHICLE 4': 'contributing_factor_vehicle_4',
    'CONTRIBUTING FACTOR VEHICLE 5': 'contributing_factor_vehicle_5',
    'COLLISION_ID': 'collision_id',
    'VEHICLE TYPE CODE 1': 'vehicle_type_code_1',
    'VEHICLE TYPE CODE 2': 'vehicle_type_code_2',
    'VEHICLE TYPE CODE 3': 'vehicle_type_code_3',
    'VEHICLE TYPE CODE 4': 'vehicle_type_code_4',
    'VEHICLE TYPE CODE 5': 'vehicle_type_code_5'
}, inplace=True)
```

```python
vehicle_collisions_df.head(1000)
```

```python
vehicle_collisions_df.isnull().sum()
```

```python
# Ensure the 'latitude' column is numeric, converting errors to NaN
vehicle_collisions_df['latitude'] = pd.to_numeric(vehicle_collisions_df['latitude'], errors='coerce')
# Group by 'LocationDesc' and 'TopicDesc', then apply forward fill and backward fill
vehicle_collisions_df['latitude'] = vehicle_collisions_df.groupby(['collision_id', 'on_street_name'])['latitude'].transform(lambda x: x.ffill().bfill())
# Check remaining missing values
remaining_missing = vehicle_collisions_df['latitude'].isnull().sum()
print("Remaining missing values in 'latitude':", remaining_missing)
# If there are still missing values, fill them with the mean of 'Data_Value'
if remaining_missing > 0:
    mean_value = vehicle_collisions_df['latitude'].mean()
    vehicle_collisions_df['latitude'] = vehicle_collisions_df['latitude'].fillna(mean_value)
# Verify missing values after all processing
print("Missing values in 'latitude' after final filling:", vehicle_collisions_df['latitude'].isnull().sum())
# Save the updated DataFrame to a new CSV file
vehicle_collisions_df.to_csv('/home/oem/machine_learning_jupyter/motor_vehicle_collisions_crashes.csv', index=False)
```
